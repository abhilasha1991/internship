# Q.1.Write a python program to display all the header tags from wikipedia.org
solution:

```python
#header tags
url = 'https://en.wikipedia.org/wiki/Main_Page'
page =requests.get(url)
page
```

```python
tags = soup.find_all(['h1','h2','h3','h4','h5','h6'])
for i in tags:
    print('list all the header tags:',*tags, sep='\n\n')
```

# Q.2 Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release)and make data frame
solution:

```python
#IMBD top rated 100 movies
url ='https://www.imdb.com/chart/top/?ref_=nv_'
page = requests.get(url)
imbd = BeautifulSoup(page.content,"html.parser")
```

```python

scraped_movies=imbd.find_all('td',class_="titleColumn")
movies=[]
for movie in scraped_movies:
    movie=movie.get_text().replace('\n',"")
    movie=movie.strip(" ")
    movies.append(movie)

scraped_rating=imbd.find_all('td',class_="ratingColumn imdbRating")
ratings=[]
for rating in scraped_rating:
    rating=rating.get_text().replace('\n','')
    ratings.append(rating)
```

```python
data=pd.DataFrame()
data['movies']=movies
data['rating']=ratings
print("\nTop 100 rated movies on IMBD")
data.head(100)
```

# Q.3 Write a python program to display IMDB’s Top rated 100 Indian movies’ data (i.e. name, rating,year of release) and make data frame
sotuion:

```python
#IMBD top rated 100 indian movies
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url='https://www.imdb.com/list/ls009997493/'
page = requests.get(url)
soup = BeautifulSoup(page.content,'html.parser')
```

```python
rank=[]
scrape_rank=soup.find_all('span',class_="lister-item-index unbold text-primary")
for i in scrape_rank:
    rank.append(i.text)
 
    
scrape_title=soup.find_all('h3',class_="lister-item-header")
titles=[]
for title in scrape_title:
    title=title.get_text().replace('\n','')
    titles.append(title)


rating=[]
scrape_rating=soup.find_all('div',class_="ipl-rating-star small")
for i in scrape_rating:
    i =i.get_text().replace('\n','')
    rating.append(i)
    
```

```python
data=list(zip(rank,titles,rating))
df=pd.DataFrame(data,columns=['rank','title','rating'])
print("\nTop 100 Indian movies on IMBD")
df

```

# Q.5 Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape:
a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating
solution:
a.

```python
#top 10 odi cricket mens team
url='https://www.icc-cricket.com/rankings/mens/team-rankings/odi'
page=requests.get(url)
page
```

```python
soup=BeautifulSoup(page.content)
```

```python
rank=[]
rank_=soup.find_all('td',class_="table-body__cell table-body__cell--position u-text-right")
for i in rank_:
    rank.append(i.text)
team=[]
team_=soup.find_all('span',class_="u-hide-phablet")
for i in team_:
    team.append(i.text)
matches=[]
match_=soup.find_all('td',class_="table-body__cell u-center-text")
for i in match_:
    matches.append(i.text)
points=[]
point_=soup.find_all('td',class_="table-body__cell u-center-text")
for i in point_:
    points.append(i.text)
ratings=[]
rating_=soup.find_all('td',class_="table-body__cell u-text-right rating")
for i in rating_:
    ratings.append(i.text)
    
    
data=list(zip(rank,team,matches,points,ratings))
df=pd.DataFrame(data,columns=['rank','team','matches','points','ratings'])
print('\nTop 10 ODI teams in men’s cricket along with the records for matches, points and rating.\n')
df.head(10)
```

```python
#news details
page=requests.get("https://www.cnbc.com/world/?region=world")
soup=BeautifulSoup(page.content,"html.parser")
```

```python
headlines=[]
headline=soup.find_all('a',class_="LatestNews-headline")
for i in headline:
    headlines.append(i.text)

time=[]
time_=soup.find_all('span',class_="LatestNews-wrapper")
for i in time_:
    time.append(i.text)
    
newslink=[]
for i in soup.find_all('a', class_="LatestNews-headline"):
    newslink.append(i.get('href'))

```

```python
data=list(zip(headlines,time,newslink)) 
data
```

```python

df=pd.DataFrame(data,columns=['headlines','time','newslink'])
print("News details")
df
```

# Q.9
solution:

```python
#details from dineout.co.in
page= requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
soup=BeautifulSoup(page.content)
```

```python
restaurant=[]
rest_=soup.find_all('a',class_="restnt-name ellipsis")
for i in rest_:
    restaurant.append(i.text)


location=[]
location_=soup.find_all('div',class_="restnt-loc ellipsis")
for i in location_:
    location.append(i.text)
    
cuisine=[]  
cuisine_=soup.find_all('span',class_="double-line-ellipsis")
for i in cuisine_:
    cuisine.append(i.text)

    
rating=[]    
rating_=soup.find_all('div',class_="restnt-rating rating-4")
for i in rating_:
    rating.append(i.text)
    
image=[]
image_=soup.find_all('div',class_="img cursor")
for i in image_:
    image.append(i.get('data-link'))

data=list(zip(restaurant,cuisine,location,rating,image))
df=pd.DataFrame(data,columns=['Restaurant name','cuisine','location','rating','image'])
print("\nDetails of restaurant from dineout.co.in")
df
    
    
```

# Q.10

```python
#top publication from google scholar form
page= requests.get('https://scholar.google.com/citations?view_op=top_venues&hl=en')
soup=BeautifulSoup(page.content)
```

```python
rank=[]
rank_=soup.find_all('td',class_="gsc_mvt_p")
for i in rank_:
    rank.append(i.text)


publication=[]
publication_=soup.find_all('td',class_="gsc_mvt_t")
for i in publication_:
    publication.append(i.text)
    
h5_index=[]    
index=soup.find_all('td',class_="gsc_mvt_n")
for i in index:
    h5_index.append(i.text)
    
h5_medium=[]
medium=soup.find_all('span',class_="gs_ibl gsc_mp_anchor")
for i in medium:
    h5_medium.append(i.text)
    
    
data=list(zip(rank,publication,h5_index,h5_medium))
df=pd.DataFrame(data,columns=['Rank','Publication','h5-index','h5-medium'])
print("\nTop publication from google scholar form")
df
```
