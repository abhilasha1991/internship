```python


#python programme to scrape top 10 "Data Analyst" position in bangalore

import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.common.by import By
import time
```

```python
driver=webdriver.Chrome(r"chromedriver.exe")
driver.get('https://www.naukri.com/')
```

```python
designation=driver.find_element(By.CLASS_NAME,"suggestor-input ")
designation.send_keys('Data Analyst')
location=driver.find_element(By.XPATH,"/html/body/div[1]/div[6]/div/div/div[5]/div/div/div/input")
location.send_keys('Bangalore')
search=driver.find_element(By.CLASS_NAME,"qsbSubmit")
search.click()
```

```python
job_title=[]
title_tags=driver.find_elements(By.XPATH,'//a[@class="title fw500 ellipsis"]')
for i in title_tags[0:10]:
    title=i.text
    job_title.append(title)
    
    
job_location=[]
location=driver.find_elements(By.XPATH,'//li[@class="fleft grey-text br2 placeHolderLi location"]')
for i in location[0:10]:
    place=i.text
    job_location.append(place)
    
    
company=[]
name=driver.find_elements(By.XPATH,'//a[@class="subTitle ellipsis fleft"]')
for i in name[0:10]:
    names=i.text
    company.append(names)
    
    
experience=[]
exp=driver.find_elements(By.XPATH,'//li[@class="fleft grey-text br2 placeHolderLi experience"]')
for i in exp[0:10]:
    exps=i.text
    experience.append(exps)
    
data=list(zip(job_title,job_location,company,experience))
df=pd.DataFrame(data,columns=['job title','job location','company name','experince required'])
print("Top 10 Data analyst position in bangalore")
df
```

# Q.10
solution:

```python
driver=webdriver.Chrome(r"chromedriver.exe")
driver.get('https://www.ambitionbox.com/')
```

```python
Company=[]
title_tags=driver.find_elements(By.XPATH,"//div[contains(@class,'company-info')]")
for i in title_tags[0:20]:
    Company.append(i.text.replace("\n",""))
res=[]
for i in Company:
    if i not in res:
        res.append(i)   
exp_=[]
exp=driver.find_elements(By.XPATH,"//div[contains(@class,'sbold-list-header')]")
for i in exp[0:10]:
    exp_.append(i.text)
record=[]
rcd=driver.find_elements(By.XPATH,"//span[contains(@class,'datapoints')]")
for i in rcd:
    record.append(i.text)
minimum=[]
name=driver.find_elements(By.XPATH,"//div[contains(@class,'salary-values')]")
for i in name[0:10]:
    minimum.append(i.text)
    
    
avg=[]
av=driver.find_elements(By.XPATH,"//p[contains(@class,'averageCtc')]")
for i in av[0:10]:
    avg.append(i.text)
    
data=list(zip(res,exp_,record,minimum,avg))
df10=pd.DataFrame(data,columns=['Company name','Experience required','Total salary record','Minimum salary','Average salary'])
new=df10["Company name"].str.split('D',n=1,expand=True)
df10["Company Name"]=new[0]
df10.drop(columns=["Company name"],inplace=True)

new_=df10["Experience required"].str.split('(',n=1,expand=True)
df10["Experience Required"]=new_[0]
df10.drop(columns=["Experience required"],inplace=True)

new_df=df10["Minimum salary"].str.split('\n',n=1,expand=True)
df10["Minimum Salary"]=new_df[0]
df10["Maximum salary"]=new_df[1]
df10.drop(columns=["Minimum salary"],inplace=True)

new_colm=["Company Name","Experience Required","Total salary record","Minimum Salary","Average salary","Maximum salary"]
df10=df10.reindex(columns=new_colm)
df10

```
