###### Q.4.Write a python program to search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) onwww.flipkart.com and scrape following details for all the search results displayed on 1st page. Details to be scraped: “Brand Name”, “Smartphone name”, “Colour”, “RAM”, “Storage(ROM)”, “Primary Camera”, “Secondary Camera”, “Display Size”, “Battery Capacity”, “Price”, “Product URL”. Incase if any of the details is missing then replace it by “- “. Save your results in a dataframe and CSV

```python
driver =webdriver.Chrome(r"chromedriver.exe")
driver.get("https://www.flipkart.com")
time.sleep(5)
```

```python
search_bar=driver.find_element(By.XPATH,"//input[contains(@class,'_3704LK')]")
search_bar.send_keys('Oneplus Nord, pixel 4A')
search=driver.find_element(By.XPATH,"//button[contains(@class,'L0Z3Pu')]")
search.click()
time.sleep(5)
```

```python
name=[]
n=driver.find_elements(By.XPATH,"//div[contains(@class,'_4rR01T')]")
for i in n:
    name.append(i.text)
detail=[]
d=driver.find_elements(By.XPATH,"//ul[contains(@class,'_1xgFaf')]")
for i in d:
    detail.append(i.text)
price=[]
p=driver.find_elements(By.XPATH,"//div[contains(@class,'_30jeq3 _1_WHN1')]")
for i in p:
    price.append(i.text)
url=[]
u=driver.find_elements(By.XPATH,"//a[contains(@class,'_1fQZEK')]")
for i in u:
    url.append(i.get_attribute('href'))
    
data=list(zip(name,detail,price,url)) 
df2=pd.DataFrame(data,columns=['smartphone','details','price','url'])
new=df2["details"].str.split('\n',n=1,expand=True)
df2["Ram/Rom"]=new[0]

new=df2["details"].str.split('\n',n=2,expand=True)
df2["Display"]=new[1]

new=df2["smartphone"].str.split('(',n=1,expand=True)
df2["Smartphone"]=new[0]
new=df2["smartphone"].str.split('(',n=2,expand=True)
df2["color"]=new[1]
new=df2["color"].str.split(',',n=1,expand=True)
df2["Colour"]=new[0]
new=df2["Smartphone"].str.split('N',n=1,expand=True)
df2["Brand"]=new[0]
new=df2["Smartphone"].str.split('N',n=2,expand=True)
df2["Phone"]=new[1]
df2.drop(columns=["smartphone","details","Smartphone","color"],inplace=True)
new_colm=["Brand","Phone","Colour","Ram/Rom","Display","price","url"]
df2=df2.reindex(columns=new_colm)
df2
```

```python
csv=df2.to_csv()
csv
```

```python
driver.close()
```

### Q.7. Write a program to scrap all the available details of best gaming laptops from digit.in.

```python
driver =webdriver.Chrome(r"chromedriver.exe")
driver.get("https://digit.in.")
time.sleep(5)
```

```python
All=driver.find_element(By.XPATH,"/html/body/div[6]/div[1]/div/div/form/div/input")
All.send_keys("best gaming laptops")
```

```python
name=[]
window=[]
display=[]
processor=[]
memory=[]
laptop=driver.find_elements(By.XPATH,"//div[contains(@class,'left_side')]")
for i in laptop:
    name.append(i.text)
a=driver.find_elements(By.XPATH,"//li[contains(@title,'Windows 11 Home OS')]")
for i in a:
    window.append(i.text)
b=driver.find_elements(By.XPATH,"//div[contains(@class,'product-detail')]")
for i in b:
    display.append(i.text)

data=list(zip(name,window,display))
df7=pd.DataFrame(data,columns=['Laptop_','Window','Display_'])
new=df7["Laptop_"].str.split('\n',n=2,expand=True)
df7["Laptop"]=new[1]
new=df7["Display_"].str.split('\n',n=2,expand=True)
df7["Display"]=new[1]
new=df7["Display_"].str.split('\n',n=3,expand=True)
df7["Processor"]=new[2]
new=df7["Display_"].str.split('\n',n=4,expand=True)
df7["Memory"]=new[3]
df7.drop(columns=["Laptop_","Display_"],inplace=True)
New_colm=['Laptop','Window','Processor','Display','Memory']
df7=df7.reindex(columns=New_colm)
df7
```
